##### 3. Twitter API & JSON

```python
#create Tweepy API
consumer_key = 'YOUR CONSUMER KEY'
consumer_secret = 'YOUR CONSUMER SECRET'
access_token = 'YOUR ACCESS TOKEN'
access_secret = 'YOUR ACCESS SECRET'

auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_secret)
api = tweepy.API(auth, wait_on_rate_limit=True)

# Create list of tweet ids
tweet_ids = archive.tweet_id.values
len(tweet_ids)

# Query Twitter's API for JSON data for each tweet ID in the Twitter archive
count = 0
fails_dict = {}
start = timer()
# Save each tweet's returned JSON as a new line in a .txt file
with open('tweet_json.txt', 'w') as outfile:
    # This loop will likely take 20-30 minutes to run because of Twitter's rate limit
    for tweet_id in tweet_ids:
        count += 1
        print(str(count) + ": " + str(tweet_id))
        try:
            tweet = api.get_status(tweet_id, tweet_mode='extended')
            print("Success")
            json.dump(tweet._json, outfile)
            outfile.write('\n')
        except tweepy.TweepError as e:
            print("Fail")
            fails_dict[tweet_id] = e
            pass
end = timer()
print(end - start)
```

```python
# check how many tweet ID fails
len(fails_dict)
```

```python
# print all failed tweet ID with error  
for tweet_id, error in fails_dict.items():
    print(str(tweet_id)+' : '+str(error)) 
```

>- Some API requests fail because tweet does not exist anymore
>- the other fail because the API failed to send request. We need to send again API request for this Tweet IDs

```python
# try to send API query for tweet ID that fails with error 'Failed to send request'
count =0
fails_dict_copy = fails_dict.copy()
with open('tweet_json.txt', 'a') as outfile:
    for tweet_id, error in fails_dict_copy.items():
        if 'Failed to send request' in str(error) :
            count += 1
            print(str(count) + ": " + str(tweet_id))

            try:
                tweet = api.get_status(tweet_id, tweet_mode='extended')
                print("Success")
                json.dump(tweet._json, outfile)
                outfile.write('\n')
                del fails_dict[tweet_id]
            except tweepy.TweepError as e:
                print("Fail")
                pass
```

```python
# confirm that all failed API queries are deleted tweets
for tweet_id, error in fails_dict.items():
        print(str(tweet_id)+' : '+str(error))
```

```python
# print the number of failed queries(deleted tweets)
len(fails_dict)
```

```python
# Extract DATA from JSON file and save it in new DataFrame
df_list = []
with open('tweet_json.txt') as json_file:
    for line in json_file:
        tweet = json.loads(line)
        tweet_id = tweet['id']
        retweet_count = tweet['retweet_count']
        favorite_count = tweet['favorite_count']
        try :
            expanded_urls = tweet['extended_entities']['media'][0]['expanded_url']
        except :
            expanded_urls = None
            pass

        df_list.append({'tweet_id': tweet_id,
                        'retweet_count': retweet_count,
                        'favorite_count': favorite_count,
                        'expanded_urls': expanded_urls
                        })
            
tweets = pd.DataFrame(df_list, columns = ['tweet_id', 'retweet_count', 'favorite_count', 'expanded_urls'])
        
```
